### StanfordParser

```python
path_jar = '/Users/choiyoungrok/myproject/nlp/stanford-corenlp-full-2018-10-05/stanford-corenlp-3.9.2.jar'
path_model = '/Users/choiyoungrok/myproject/nlp/stanford-corenlp-full-2018-10-05/stanford-corenlp-3.9.2-models.jar'
parser = stanford.StanfordParser(path_to_jar=path_jar, path_to_models_jar=path_model)
```

<br>
#### parser.raw_parse_sents

```python
%%time
sentences = parser.raw_parse_sents(("Hello, My name is Steve.", "What is your name?", "I would like to have some water"))
parsed_list = []
for line in sentences:
    for val in line:
        parsed_list.append(val)
```
